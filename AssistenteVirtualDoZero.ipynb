```python
import speech_recognition as sr
from gtts import gTTS
from IPython.display import Audio, display
import wikipedia
import webbrowser
import os
```

```python
def text_to_speech(texto):
    """Converte texto em áudio"""
    tts = gTTS(text=texto, lang='pt-br', slow=False)
    tts.save('resposta.mp3')
    display(Audio('resposta.mp3', autoplay=True))
    print(f"Assistente: {texto}")
```

```python
def speech_to_text():
    """Reconhece fala e converte em texto"""
    recognizer = sr.Recognizer()

    # Para upload de arquivo de áudio
    from google.colab import files
    uploaded = files.upload()

    audio_file = list(uploaded.keys())[0]

    with sr.AudioFile(audio_file) as source:
        audio_data = recognizer.record(source)
        try:
            texto = recognizer.recognize_google(audio_data, language='pt-BR')
            return texto.lower()
        except:
            return ""
```

```python
def executar_comando(comando):
    """Executa ações baseadas no comando"""

    if 'wikipedia' in comando or 'wikipédia' in comando:
        wikipedia.set_lang('pt')
        termo = comando.replace('wikipedia', '').replace('wikipédia', '').strip()
        resultado = wikipedia.summary(termo, sentences=2)
        text_to_speech(resultado)

    elif 'youtube' in comando:
        webbrowser.open('https://www.youtube.com')
        text_to_speech('Abrindo YouTube')

    elif 'farmácia' in comando or 'farmacia' in comando:
        webbrowser.open('https://www.google.com/maps/search/farmácia')
        text_to_speech('Mostrando farmácias próximas')

    else:
        text_to_speech('Comando não reconhecido')
```

```python
def executar_comando(comando):
    """Executa ações baseadas no comando"""

    if 'wikipedia' in comando or 'wikipédia' in comando:
        try:
            wikipedia.set_lang('pt')
            termo = comando.replace('wikipedia', '').replace('wikipédia', '').replace('pesquisar', '').replace('na', '').strip()

            # Adicione auto_suggest=False para evitar autocorreção incorreta
            resultado = wikipedia.summary(termo, sentences=2, auto_suggest=False)
            text_to_speech(resultado)
            print(f"Resumo: {resultado}")

        except wikipedia.exceptions.DisambiguationError as e:
            # Página de desambiguação - pega primeira opção
            text_to_speech(f'Encontrei várias opções. Buscando {e.options[0]}')
            resultado = wikipedia.summary(e.options[0], sentences=2)
            text_to_speech(resultado)
            print(f"Resumo: {resultado}")

        except wikipedia.exceptions.PageError:
            # Página não existe
            text_to_speech('Não encontrei essa página na Wikipédia')
            print("PageError: Página não encontrada")

        except Exception as e:
            text_to_speech('Erro ao buscar informação')
            print(f"Erro: {e}")

    elif 'youtube' in comando:
        webbrowser.open('https://www.youtube.com')
        text_to_speech('Abrindo YouTube')

    elif 'farmácia' in comando or 'farmacia' in comando:
        webbrowser.open('https://www.google.com/maps/search/farmácia')
        text_to_speech('Mostrando farmácias próximas')

    else:
        text_to_speech('Comando não reconhecido')
```

```python
def testar_comandos():
    """Testa comandos sem precisar de áudio"""
    comandos_teste = [
        'pesquisar python na wikipédia',
        'pesquisar Brasil na wikipédia',
    ]

    for comando in comandos_teste:
        print(f"\n{'='*50}")
        print(f"Testando: {comando}")
        print(f"{'='*50}")
        executar_comando(comando)

testar_comandos()
```
